```python
df = df_marge.drop(["title", "subject", "date"], axis=1, errors='ignore')
```

```python
df = df.sample(frac = 1)
```

```python
df.reset_index(inplace=True)
df.drop(["index"], axis=1, inplace=True, errors='ignore')
```

```python
def wordopt(text):
    # Convert to lowercase
    text = text.lower()

    # Remove content inside square brackets (optional: modify regex for specific cases)
    text = re.sub(r'\[.*?\]', '', text)
    
    # Replace non-word characters (including punctuation) with a space
    text = re.sub(r'\W+', ' ', text)

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # Remove punctuation (optional: already handled by \W+)
    # text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    
    # Remove newline characters
    text = re.sub(r'\n', '', text)

    # Remove words with numbers (use carefully for specific cases)
    text = re.sub(r'\w*\d\w*', '', text)

    # Remove extra spaces (if any)
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df["text"] = df["text"].apply(wordopt)
```

```python
x = df["text"]
y = df["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(xv_train, y_train)
```

```python
pred_gbc = GBC.predict(xv_test)
```

```python
GBC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)
```

```python
pred_rfc = RFC.predict(xv_test)
```

```python
RFC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "True News"
    else:
        return "Unknown"

def manual_testing(news):
    # Ensure the input is a string and not empty
    if not isinstance(news, str) or not news.strip():
        raise ValueError("Input must be a non-empty string.")

    # Create DataFrame for the input news
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)

    # Preprocess the text
    new_def_test["text"] = new_def_test["text"].apply(wordopt) 
    new_x_test = new_def_test["text"]
    
    # Vectorize the input
    new_xv_test = vectorization.transform(new_x_test)
    
    # Make predictions using different models
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)
    

    # Prepare results in a structured format
    results = {
        "Logistic Regression": output_label(pred_LR[0]),
        "Decision Tree": output_label(pred_DT[0]),
        "Gradient Boosting": output_label(pred_GBC[0]),
        "Random Forest": output_label(pred_RFC[0])
    }

    return results

# Get user input
user_input = input("Enter the news article to check: ")
try:
    predictions = manual_testing(user_input)

    # Print the predictions
    print("\nPredictions:")
    for model, result in predictions.items():
        print(f"{model}: {result}")
except ValueError as e:
    print(e)
```

```python
import pickle

# Save the models
with open('LR.pkl', 'wb') as f:
    pickle.dump(LR, f)
with open('DT.pkl', 'wb') as f:
    pickle.dump(DT, f)
with open('GBC.pkl', 'wb') as f:
    pickle.dump(GBC, f)
with open('RFC.pkl', 'wb') as f:
    pickle.dump(RFC, f)

# Save the vectorizer
with open('vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorization, f)
```
